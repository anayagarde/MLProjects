# Model

```python
x = df.drop('price',axis=1).values # x is all columns exept price (input features - 19)
y = df['price'].values   # y label is price
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
print('X_train shape: ', x_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', x_test.shape)
print('y_test shape: ', y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
print("Tensorflow Version",tf.__version__)
```

```python
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras import Sequential
```

**NOTE**     
**Model Architecture**         
***6 layer Neural Network***   
A. Dense Layer 1 - 32 neurons         
B. Dense Layer 2 - 19 neurons          
C. Dense Layer 3, 4 and 5 - 20 neurons             
D. Dense Layer 6 - 1 neuron (output Layer)           

```python
model = Sequential()
    
model.add(Dense(32,activation='relu',input_shape=(19,))),
model.add(Dense(19,activation='relu')),
model.add(Dense(20,activation='relu')),
model.add(Dense(20,activation='relu')),
model.add(Dense(20,activation='relu')),
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss=tf.keras.losses.MeanSquaredError()) #Optimization=adam and loss function = mean squared error
```

```python
model.fit(x_train,y_train,
          batch_size=64,
          epochs = 200,
          validation_data=(x_test,y_test))
```

```python
model.summary()
```

```python
loss = pd.DataFrame(model.history.history)
loss.plot(figsize=(10,8))
```

## Predictions 

```python
pred = model.predict(x_test)
```

```python
from sklearn import metrics
print('Mean Squared Error:', metrics.mean_squared_error(y_test, pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, pred))  
print('Variance score:',metrics.explained_variance_score(y_test,pred)*100)
#ideally variance score should not be less than 60
```

```python
sns.set_style('whitegrid')
fig = plt.figure(figsize=(15,8))
plt.scatter(y_test,pred,edgecolors='None') 
plt.plot(y_test,y_test,'g') #green line indicates y_test i.e.label of test set -> perfect prediction
```

```python
plt.figure(figsize=(8,5))
diff = (y_test - pred)
sns.distplot(diff,color='red')
```
